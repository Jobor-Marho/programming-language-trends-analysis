# Programming Language Trends Analysis
This project analyzes trends in programming language discussions using Stack Overflow post data.
The goal is to understand how developer interest in different programming languages has changed over time.

## Project Objective
- Examine historical trends in programming language popularity
- Identify languages gaining or losing traction
- Compare growth patterns across major programming technologies
- Visualize developer interest over time

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

pd.options.display.float_format = '{:,.2f}'.format
```

## Load Data

```python
reshaped_df = pd.read_csv('../data/reshaped_df.csv')
rolling_df = pd.read_csv('../data/rolling_df.csv')

reshaped_df.head()
```

## Data Understanding

```python
reshaped_df.shape
reshaped_df.columns
reshaped_df.info()
```

## Data Cleaning

```python
reshaped_df["DATE"] = pd.to_datetime(reshaped_df["DATE"]).dt.normalize()
rolling_df["DATE"] = pd.to_datetime(rolling_df["DATE"]).dt.normalize()
```

## Trend Visualization

```python
plt.figure(figsize=(16,10))

plt.plot(reshaped_df['DATE'], reshaped_df['python'], label='Python')
plt.plot(reshaped_df['DATE'], reshaped_df['java'], label='Java')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Posts',fontsize=14)
plt.ylim((0, 35000))
plt.legend(fontsize=18)
plt.show()
```

## Trend Smoothing (Rolling Average)
A rolling average helps reduce short-term fluctuations and highlight long-term trends.

```python
rolling_df.head()
```

## Interactive Visualization

```python
fig = px.line(rolling_df, x='DATE', y=rolling_df.columns[1:], title='Programming Language Popularity Trends')
fig.show()
```

## Key Insights
- Python shows strong growth in developer discussions over time.
- Established languages such as Java maintain steady engagement across years.
- Some programming languages experience rapid increases in popularity before stabilizing.
- The data highlights how developer interest evolves as new technologies emerge.
- Trend analysis of developer discussions can help identify emerging technologies.

## Conclusion
This analysis highlights how interest in programming languages evolves within the developer community.
Understanding these trends can help developers, organizations, and educators identify which technologies are gaining traction and where future opportunities may lie.
